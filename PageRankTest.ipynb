```python
# Algorithm execition using python library

import networkx as nx

G = nx.DiGraph()

G.add_edge('A', 'E')
G.add_edge('B', 'A')
G.add_edge('B', 'E')
G.add_edge('C', 'A')
G.add_edge('C', 'D')
G.add_edge('D', 'A')
G.add_edge('D', 'B')
G.add_edge('D', 'C')
G.add_edge('D', 'E')
G.add_edge('E', 'A')

d = 0.85

pagerank = nx.pagerank(G, alpha=d)
pagerank_sorted = dict(sorted(pagerank.items()))

print(pagerank_sorted)
```

```python
import matplotlib.pyplot as plt
nx.draw(G, with_labels=True)
```
